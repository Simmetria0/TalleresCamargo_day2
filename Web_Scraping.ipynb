```python
#import libraries
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
```

```python
#request access
url = "http://web.mta.info/developers/"
#url = "http://web.mta.info/developers/turnstile.html"
response = requests.get(url+"turnstile.html")

```

```python
#check if everything went fine
print(response)
```

```python
soup = BeautifulSoup(response.text, "html.parser")
```

```python
#ahora buscamos todos los <a> tags:
soup.findAll('a')

#que es el <a> tag? 
#https://www.w3schools.com/tags/tag_a.asp
```

```python
# y decidimos que queremos el hiperlink en la linea 36
one_a_tag = soup.findAll("a")[36]
link = one_a_tag["href"]
```

```python
#and we can download it
download_url = url+ link
```

```python
a= urllib.request.urlretrieve(download_url,"./"+link[link.find("/turnstile_")+1:])
```

```python
#pausa para que no salten alarmas
time.sleep(1)
```

```python
# y ahora que sabemos como se descargan...
import os
#os.mkdir('DatabaseScraping')
#os.chdir("./DatabaseScraping")


for i in range(36,48):#len(soup.findAll('a'))+1): 
    one_a_tag = soup.findAll('a')[i]
    link = one_a_tag['href']
    download_url = url+ link
    urllib.request.urlretrieve(download_url,'./'+link[link.find('/turnstile_')+1:]) 
    time.sleep(1) #pause the code for a sec
```
